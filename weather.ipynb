# The Weather Dataset


```python
import pandas as pd
data = pd.read_csv('weather.csv')
data.head(3)
#data.shape
```

```python
data.shape
```

```python
# Find Missing values in the data set
data.isnull().sum()
```

```python
# Fill missing values
data.fillna(method='ffill', inplace=True)
data.head(10)
```

```python
# Rename the Column Name?
data.rename(columns={'Temp':'Temperature'}, inplace=True)
```

```python
data
```

```python
# Find All the Unique “Wind Speed” Values in the Data?
data.nunique()
```

```python
data['Wind Speed'].nunique()
```

```python
data['Wind Speed'].unique()
```

```python
# Find the Number of Times when the “Weather is Exactly Clear”?
data.Weather.value_counts()
```

```python
data[data.Weather=='Clear']
```

```python
data.groupby('Weather').get_group('Clear')
```

```python
# When Wind Speed Was Exactly 4 km/h?
data[data['Wind Speed']==4]
```

```python
# Mean Visibility
data.Visibility.mean()
```

```python
# Standard Deviation of Pressure?
data.Pressure.std()
```

```python
# Variance of Relative Humidity?
data['Relative Humidity'].var()
```

```python
# All the Instances When Snow is Covered?
data['Weather'].value_counts()
```

```python
data[data['Weather']=='Snow']
```

```python
data[data['Weather'].str.contains('Snow')]
```

```python
# All the Instances When Wind Speed is Above 24 and Visibility is 25?
data[(data["Wind Speed"]>24) & (data["Visibility"]==25)]
```

```python
# Mean Value of Each Column Against Each Weather Condition?
# Filter only numeric columns
numeric_df = data.select_dtypes(include='number')
numeric_df["Weather"]=data['Weather']
numeric_df.groupby("Weather").mean()
```

```python
# What is the Maximum and Minimum Value of Each Column Against Each Weather?
data.groupby('Weather').min()
```

```python
# Show All the Records Where Weather Condition is Fog?
data[data['Weather']=='Fog']
```

```python
# Find All the Instances When Weather is Clear or Visibility is Above 40
data[(data['Weather']=='Clear') | (data['Visibility'] > 40)]
```

```python
# Find All the Instances When “Weather is Clear and Relative Humidity is Greater than 50” or Visibility is Above 40
data[(data["Weather"]=='Clear') & (data["Relative Humidity"]>50) | (data["Visibility"]>40)]
```
